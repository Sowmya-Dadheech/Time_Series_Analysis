# **Applying Filter**

```python
df[(df['Country']=='Denmark') & (df['Smart_Living']>=5000)]
```

# **Unique Values**

```python
df['Id'].unique()
```

```python
Query_string = """ select distinct Id from dfpsql """
sqldf(Query_string, globals())
```

# IN

```python
df[df['Country'].isin(['Norway','Russia'])]
```

```python
Query_string = """ select * from dfpsql where Country in('Norway','Russia')"""
sqldf(Query_string, globals())
```

# NOT IN

```python
df[~df['Country'].isin(['Norway','Russia'])]
```

```python
Query_string = """ select * from dfpsql where Country not in('Norway','Russia')"""
sqldf(Query_string, globals())
```

# Ascending Data order

```python
df.sort_values(by = ['City','SmartCity_Index'], ascending= True)
```

```python
Query_string = """ select * from dfpsql order by City,SmartCity_Index;"""
sqldf(Query_string, globals())
```

# Descending Data Order

```python
df.sort_values(by = ['City','SmartCity_Index'], ascending= False)
```

```python
Query_string = """ select * from dfpsql order by City,SmartCity_Index Desc;"""
sqldf(Query_string, globals())
```

# Aggregation

```python
df.agg({'Smart_People': ['count','min', 'max', 'mean']})
```

```python
Query_string = """ select count('Smart_People') as count,
min('Smart_People') as min, max('Smart_People') as max,
avg('Smart_People') as mean from df;"""
sqldf(Query_string, globals())
```

# Group By

```python
df.groupby('Country')['Smart_People'].sum()
```

```python
Query_string = """ select Country , sum(Smart_People) as Sum_Smart_People from df group by Country;"""
sqldf(Query_string, globals())
```

# Group By with Aggregation

```python
df.groupby('SmartCity_Index_relative_Edmonton').agg({'Smart_People': ['mean','min','max']})
```

```python
Query_string = """ select SmartCity_Index_relative_Edmonton , avg(Smart_People) as mean, min(Smart_People) as min, max(Smart_People) as max from df group by SmartCity_Index_relative_Edmonton;"""
sqldf(Query_string, globals())
```

# Join (Merge)

```python
import pandas as pd
data1 = { 'Empid': [1011, 1012, 1013, 1014, 1015],
         'Name': ['John', 'Rahul', 'Rick', 'Morty', 'Tim'],
          'Designation': ['Manager', 'Research Engineer', ' Research Engineer', 'VP', 'Delivery Manager'],
          'Date_of_joining': ['01-Jan-2000', '23-sep-2006', '11-Jan-2012', '21-Jan-1991', '12-Jan-1990']}
Emp_df = pd.DataFrame(data1, columns = ['Empid', 'Name', 'Designation','Date_of_joining'])
Emp_df
```

```python
data2 = { 'Empid': [1011, 1017, 1013, 1019, 1015],
         'Deptartment': ['Management', 'Research', 'Research', 'Management', 'Delivery'],
          'Total_Experience': [18, 10, 10, 28, 22]}
Dept_df = pd.DataFrame(data2, columns = ['Empid', 'Deptartment', 'Total_Experience'])
Dept_df
```

Inner Join

```python
pd.merge(Emp_df, Dept_df, left_on='Empid',right_on='Empid', how='inner')
```

```python
Query_string = """ select * from Emp_df a INNER JOIN Dept_df b ON a.Empid = b.Empid;"""
sqldf(Query_string, globals())
```

Left Join

```python
pd.merge(Emp_df, Dept_df,left_on='Empid',right_on='Empid', how='left')
```

```python
Query_string = """ select * from Emp_df a LEFT JOIN Dept_df b ON a.Empid = b.Empid;"""
sqldf(Query_string, globals())
```

Right Join

```python
pd.merge(Emp_df, Dept_df,left_on='Empid',right_on='Empid', how='right')
```

```python
Query_string = """ select a.Empid,Name,Designation,Date_of_joining,Deptartment,Total_Experience from Dept_df a LEFT JOIN Emp_df b ON a.Empid = b.Empid;"""
sqldf(Query_string, globals())
```

Outer Join

```python
pd.merge(Emp_df, Dept_df,left_on='Empid',right_on='Empid', how='outer')
```

```python
Query_string = """ select * from Emp_df a left OUTER JOIN Dept_df b ON a.Empid = b.Empid;"""
sqldf(Query_string, globals())
```
